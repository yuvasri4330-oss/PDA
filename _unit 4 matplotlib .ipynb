```python

import matplotlib.pyplot as plt 
x = [0,2,4,6,8]
y = [0,4,16,36,64]

fig ,ax = plt.subplots()
ax.plot(x,y,marker='o', label ="data points")

ax.set_title("basic components of matplotlip figure")
ax.set_xlabel("x-axis")
ax.set_xlabel("y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt 

x=['thur','fri','sar','sun']
y=[178,120,250,190]

plt.bar(x,y)
plt.title("bar yuva chart")
plt.xlabel("day")
plt.ylabel("total bill")
plt.show()
```

```python
#line plot 
import matplotlib.pyplot as plt


x=[1,2,3,4]
y=[10,20,30,40]

plt.plot(x,y)
plt.title("line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#bar chart 
import matplotlib.pyplot as plt 

categories=['a','b','c','d']
values=[5,7,3,8]

plt.bar(categories,values)
plt.title("bar chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
#HISTOGRAM
import matplotlib.pyplot as plt

data=[1,2,2,3,3,4,4,4]

plt.hist(data,bins=4)
plt.title("HISTOGRAM")
plt.xlabel("value")
plt.ylabel("frequence")
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[5,7,6,8,7]
plt.scatter(x,y)
plt.title("SCATTER")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#PIE chart
import matplotlib.pyplot as plt

labels=['a','b,','c','d']
sizes=[20,30,25,25]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("PIE CHART")
plt.show()

```

```python
#STACKED BAR CHART
import matplotlib.pyplot as plt

x=['a','b','c']
y1=[3,5,2]
y2=[4,5,3]

plt.bar(x,y1,label='data1')
plt.bar(x,y2,label='data2')

plt.title("STACKED BAR CHART")
plt.legend()
plt.show()


```

```python
#step plot
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,30,40]

plt.step(x,y)
plt.title("STEP PLOT")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

```

```python
#area plt
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,15,25]

plt.fill_between(x,y)
plt.title("AREA PLOT")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()



```

```python
#box plot
import matplotlib.pyplot as plt
data=[10,20,30,40,50,100]
plt.boxplot(data)
plt.title("box PLOT")
plt.show()


```

```python
#STEM PLOT
import matplotlib.pyplot as plt

x=[1,2,3,4,]
y=[5,15,10,20]

plt.stem(x,y)
plt.title("STEM PLOT")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
